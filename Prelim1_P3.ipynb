```python
import scipy.optimize as opt
from numpy import exp

def f(var) :
    kd,w1,w2,k2,n2 = var
    f1 = 19 - 20143*2/(kd+2)*(w1+w2*(0/k2)**n2/(1+(0/k2)**n2))/(1+w1+w2*(0/k2)**n2/(1+(0/k2)**n2))
    f2 = 21 - 20143*2/(kd+2)*(w1+w2*(5e-4/k2)**n2/(1+(5e-4/k2)**n2))/(1+w1+w2*(5e-4/k2)**n2/(1+(5e-4/k2)**n2))
    f3 = 41 - 20143*2/(kd+2)*(w1+w2*(0.005/k2)**n2/(1+(0.005/k2)**n2))/(1+w1+w2*(0.005/k2)**n2/(1+(0.005/k2)**n2))
    f4 = 67 - 20143*2/(kd+2)*(w1+w2*(0.012/k2)**n2/(1+(0.012/k2)**n2))/(1+w1+w2*(0.012/k2)**n2/(1+(0.012/k2)**n2))
    f5 = 86 - 20143*2/(kd+2)*(w1+w2*(0.053/k2)**n2/(1+(0.053/k2)**n2))/(1+w1+w2*(0.053/k2)**n2/(1+(0.053/k2)**n2))
    f6 = 93 - 20143*2/(kd+2)*(w1+w2*(0.216/k2)**n2/(1+(0.216/k2)**n2))/(1+w1+w2*(0.216/k2)**n2/(1+(0.216/k2)**n2))
    f7 = 93 - 20143*2/(kd+2)*(w1+w2*(1/k2)**n2/(1+(1/k2)**n2))/(1+w1+w2*(1/k2)**n2/(1+(1/k2)**n2))
    
    return [f1,f2,f3,f4,f5,f6,f7]

sol = opt.leastsq(f, (300, 1, 1, 1, 1))[0]
print(sol)
```
