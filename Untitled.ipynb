### Problem 2 Part(b)

```python
import scipy.optimize as opt
from numpy import exp

def f(var) :
    w1,w2,k2,n2 = var
    f1 = 3 - 69.58*(w1/(1+w1))
    f2 = 6.3 - 69.58*(w1+w2*(0.055/k2)**n2/(1+(0.055/k2)**n2))/(1+w1+w2*(0.055/k2)**n2/(1+(0.055/k2)**n2))
    f3 = 29.8 - 69.58*(w1+w2*(0.093/k2)**n2/(1+(0.093/k2)**n2))/(1+w1+w2*(0.093/k2)**n2/(1+(0.093/k2)**n2))
    f4 = 52.0 - 69.58*(w1+w2*(0.181/k2)**n2/(1+(0.181/k2)**n2))/(1+w1+w2*(0.181/k2)**n2/(1+(0.181/k2)**n2))
    f5 = 60.3 - 69.58*(w1+w2*(0.405/k2)**n2/(1+(0.405/k2)**n2))/(1+w1+w2*(0.405/k2)**n2/(1+(0.405/k2)**n2))
    f6 = 68.7 - 69.58*(w1+w2*(0.99/k2)**n2/(1+(0.99/k2)**n2))/(1+w1+w2*(0.99/k2)**n2/(1+(0.99/k2)**n2))
    f7 = 68.9 - 69.58*(w1+w2*(1/k2)**n2/(1+(1/k2)**n2))/(1+w1+w2*(1/k2)**n2/(1+(1/k2)**n2))
    
    return [f1,f2,f3,f4,f5,f6]

sol = opt.leastsq(f, (1, 1, 1, 1))[0]
print(sol)
```
